```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Sales_Details.csv")
print("First 5 Rows")
df
```

```python
print("\nDataset Information")
print(df.info())
```

```python
df.isnull()
```

```python
print("\nMissing Values")
print(df.isnull().sum())
```

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mode()[0])
df.head(10)
```

```python
df["Price"] = df["Price"].fillna(int(df["Price"].mean()))
df.head(10)
```

```python
df["Customer City"] = df["Customer City"].fillna(df["Customer City"].mode()[0])
df.head(10)
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]
df.head(10)
```

```python
df.isnull().sum()
```

```python
top_products = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
print(top_products)
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"], format="%d/%m/%Y")
df["Month"]=df["Order_Date"].dt.month_name()
```

```python
monthly_sales = df.groupby("Month")["Revenue"].sum()
print(monthly_sales)
```

```python
print("Highest Sale Month")
print(monthly_sales.idxmax())
```

```python
print("Lowset Sale Month")
print(monthly_sales.idxmin())
```

```python
print("Highest Revenue Amount")
print(monthly_sales.max())
```

```python
print("Lowest Revenue Amount")
print(monthly_sales.min())
```

```python
city_sales=df.groupby("Customer City")["Revenue"].sum()
print(city_sales)
```

```python
product_qty = df.groupby("Product")["Quantity"].sum()
product_qty.plot(kind="bar")
plt.title("Total Quantity Sold by Product")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```

```python
import matplotlib.pyplot as plt
top_products.plot(kind="bar")
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
category_sales = df.groupby("Category")["Revenue"].sum()
category_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Revenue by Category")
plt.show()
```

```python
plt.hist(df["Revenue"], bins=10)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df["Revenue"])

plt.title("Revenue Distribution")

plt.show()
```
